```python
url = 'https://www.investing.com/instruments/HistoricalDataAjax'
head = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Safari/537.36','X-Requested-With': 'XMLHttpRequest' }
form = {'action': "historical_data",
       'curr_id': "8849",
       'st_date' : "01/01/2020",
       'end_date' : "10/31/2020",
       'interval_sec' : "Monthly"
       }
data = req.post(url, headers = head,data = form)
data.status_code
```

```python
soup = BeautifulSoup(data.text)
```

```python
oildata = pd.read_html(str(soup.find('table')))[0]
```
